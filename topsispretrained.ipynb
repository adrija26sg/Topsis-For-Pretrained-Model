```python
import time
import numpy as np
import pandas as pd
import torch
import evaluate
import matplotlib.pyplot as plt

from datasets import load_dataset
from transformers import pipeline, AutoModelForSeq2SeqLM, AutoTokenizer, AutoConfig

# ----------------------------
# 1. Define the models to evaluate
# ----------------------------
MODEL_DICT = {
    "BART": "facebook/bart-large-xsum",
    "PEGASUS": "google/pegasus-xsum",
    "T5": "t5-base",
    "FLAN-T5": "google/flan-t5-large",
    "LongT5": "google/long-t5-local-base"
}

# ----------------------------
# 2. Evaluation settings and TOPSIS weights
# ----------------------------
CRITERIA_WEIGHTS = {
    "rouge1": 0.30,
    "rouge2": 0.20,
    "rougeL": 0.30,
    "inference_time": 0.10,
    "model_size": 0.10
}

CRITERIA_TYPE = {
    "rouge1": True,
    "rouge2": True,
    "rougeL": True,
    "inference_time": False,
    "model_size": False
}

# ----------------------------
# 3. Load XSum dataset (small subset for testing)
# ----------------------------
def load_xsum_sample(num_samples=10, split="test"):
    dataset = load_dataset("xsum", split=split)
    dataset = dataset.shuffle(seed=42).select(range(num_samples))
    return dataset

# ----------------------------
# 4. Evaluate a single model
# ----------------------------
def evaluate_model(model_name, dataset):
    print(f"\nEvaluating model: {model_name}")
    summarizer = pipeline("summarization", model=model_name, tokenizer=model_name, device=0 if torch.cuda.is_available() else -1)

    config = AutoConfig.from_pretrained(model_name)
    model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
    model_size_millions = sum(p.numel() for p in model.parameters()) / 1e6
    print(f"Model size: {model_size_millions:.2f} million parameters")

    rouge = evaluate.load("rouge")

    rouge1_scores, rouge2_scores, rougeL_scores, inference_times = [], [], [], []

    for i, sample in enumerate(dataset):
        document = sample["document"]
        reference_summary = sample["summary"]

        start_time = time.time()
        try:
            output = summarizer(document, max_length=60, min_length=10, do_sample=False)
        except Exception as e:
            print(f"Error: {e}")
            continue
        end_time = time.time()

        inference_times.append(end_time - start_time)
        generated_summary = output[0]["summary_text"]

        rouge_result = rouge.compute(predictions=[generated_summary], references=[reference_summary])

        rouge1_scores.append(rouge_result["rouge1"])
        rouge2_scores.append(rouge_result["rouge2"])
        rougeL_scores.append(rouge_result["rougeL"])

        print(f"Sample {i+1}: Time={inference_times[-1]:.2f}s, R1={rouge1_scores[-1]:.3f}, R2={rouge2_scores[-1]:.3f}, RL={rougeL_scores[-1]:.3f}")

    return {
        "rouge1": np.mean(rouge1_scores),
        "rouge2": np.mean(rouge2_scores),
        "rougeL": np.mean(rougeL_scores),
        "inference_time": np.mean(inference_times),
        "model_size": model_size_millions
    }

# ----------------------------
# 5. TOPSIS Implementation
# ----------------------------
def topsis(decision_matrix, weights, criteria_type):
    norm_matrix = decision_matrix.copy()
    for col in decision_matrix.columns:
        norm_matrix[col] = decision_matrix[col] / np.sqrt((decision_matrix[col]**2).sum())

    weighted_matrix = norm_matrix.copy()
    for col in weighted_matrix.columns:
        weighted_matrix[col] *= weights[col]

    ideal_best, ideal_worst = {}, {}
    for col in weighted_matrix.columns:
        if criteria_type[col]:
            ideal_best[col] = weighted_matrix[col].max()
            ideal_worst[col] = weighted_matrix[col].min()
        else:
            ideal_best[col] = weighted_matrix[col].min()
            ideal_worst[col] = weighted_matrix[col].max()

    distances_best, distances_worst = [], []
    for _, row in weighted_matrix.iterrows():
        distances_best.append(np.sqrt(sum((row[col] - ideal_best[col])**2 for col in weighted_matrix.columns)))
        distances_worst.append(np.sqrt(sum((row[col] - ideal_worst[col])**2 for col in weighted_matrix.columns)))

    return np.array(distances_worst) / (np.array(distances_best) + np.array(distances_worst))

# ----------------------------
# 6. Main Execution
# ----------------------------
def main():
    dataset = load_xsum_sample(num_samples=10)
    results = {model_alias: evaluate_model(model_name, dataset) for model_alias, model_name in MODEL_DICT.items()}

    df = pd.DataFrame(results).T
    print("\nEvaluation Metrics:")
    print(df)

    df["topsis_score"] = topsis(df, CRITERIA_WEIGHTS, CRITERIA_TYPE)
    df["rank"] = df["topsis_score"].rank(ascending=False)
    df_sorted = df.sort_values(by="topsis_score", ascending=False)

    print("\nTOPSIS Ranking:")
    print(df_sorted[["topsis_score", "rank"]])

    # ----------------------------
    # 7. Visualizations
    # ----------------------------

    # 🔹 1. ROUGE-1, ROUGE-2, ROUGE-L Comparison
    df_sorted[["rouge1", "rouge2", "rougeL"]].plot(kind='bar', figsize=(10, 6), colormap='coolwarm')
    plt.xlabel("Model")
    plt.ylabel("ROUGE Score")
    plt.title("ROUGE Score Comparison")
    plt.xticks(rotation=45)
    plt.legend(["ROUGE-1", "ROUGE-2", "ROUGE-L"])
    plt.tight_layout()
    plt.savefig("rouge_comparison.png")
    plt.show()

    # 🔹 2. Efficiency Metrics (Inference Time & Model Size)
    fig, ax1 = plt.subplots(figsize=(10, 6))

    ax1.set_xlabel("Model")
    ax1.set_ylabel("Inference Time (sec)", color='tab:blue')
    ax1.bar(df_sorted.index, df_sorted["inference_time"], color='tab:blue', alpha=0.6, label="Inference Time")
    ax1.tick_params(axis='y', labelcolor='tab:blue')

    ax2 = ax1.twinx()
    ax2.set_ylabel("Model Size (Million Params)", color='tab:red')
    ax2.plot(df_sorted.index, df_sorted["model_size"], marker='o', color='tab:red', label="Model Size")
    ax2.tick_params(axis='y', labelcolor='tab:red')

    plt.title("Efficiency Metrics Comparison")
    fig.tight_layout()
    plt.savefig("efficiency_metrics.png")
    plt.show()

if __name__ == "__main__":
    main()
```
